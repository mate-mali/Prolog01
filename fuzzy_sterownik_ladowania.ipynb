# Sterownik Ładowania Baterii - Logika Rozmyta (Mamdani)

Solarny sterownik ładowania baterii oparty na systemie rozmytym **Mamdani** z trapezoidalnymi funkcjami przynależności.

## Zmienne wejściowe
| Zmienna | Opis | Zakres |
|---------|------|--------|
| `SOC` | poziom naładowania baterii | 0 – 100 % |
| `temp_baterii` | temperatura baterii | 0 – 60 °C |
| `temp_sterownika` | temperatura sterownika ładowania | 0 – 80 °C |
| `prad_rozladowania` | pobierany prąd / prąd rozładowania | 0 – 100 A |
| `napiecie_paneli` | napięcie na panelach solarnych | 0 – 48 V |

## Zmienna wyjściowa
| Zmienna | Opis | Zakres |
|---------|------|--------|
| `prad_ladowania` | docelowy prąd ładowania baterii | 0 – 60 A |

Każda zmienna dzielona jest na 5 klas rozmytych trapezoidalnych.

```python
from simpful import *

FS = FuzzySystem()
```

## 1. SOC -Poziom naładowania baterii [0–100 %]

Klasy: bardzo_niski, niski, sredni, wysoki, bardzo_wysoki

```python
t1 = TrapezoidFuzzySet(0,   0,  15,  25, term="bardzo_niski")
t2 = TrapezoidFuzzySet(15, 25,  35,  45, term="niski")
t3 = TrapezoidFuzzySet(35, 45,  55,  65, term="sredni")
t4 = TrapezoidFuzzySet(55, 65,  75,  85, term="wysoki")
t5 = TrapezoidFuzzySet(75, 85, 100, 100, term="bardzo_wysoki")

FS.add_linguistic_variable("SOC",
    LinguisticVariable([t1, t2, t3, t4, t5], universe_of_discourse=[0, 100]))
FS.plot_variable("SOC")
```

## 2. temp_baterii - Temperatura baterii [0–60 °C]

Klasy: bardzo_niska, niska, srednia, wysoka, bardzo_wysoka

```python
t1 = TrapezoidFuzzySet(0,   0,   9,  15, term="bardzo_niska")
t2 = TrapezoidFuzzySet(9,  15,  21,  27, term="niska")
t3 = TrapezoidFuzzySet(21, 27,  33,  39, term="srednia")
t4 = TrapezoidFuzzySet(33, 39,  45,  51, term="wysoka")
t5 = TrapezoidFuzzySet(45, 51,  60,  60, term="bardzo_wysoka")

FS.add_linguistic_variable("temp_baterii",
    LinguisticVariable([t1, t2, t3, t4, t5], universe_of_discourse=[0, 60]))
FS.plot_variable("temp_baterii")
```

Klasy: bardzo_niska, niska, srednia, wysoka, bardzo_wysoka

## 4. prad_rozladowania - Prąd rozładowania [0–100 A]

```python
t1 = TrapezoidFuzzySet(0,   0,  15,  25, term="bardzo_niski")
t2 = TrapezoidFuzzySet(15, 25,  35,  45, term="niski")
t3 = TrapezoidFuzzySet(35, 45,  55,  65, term="sredni")
t4 = TrapezoidFuzzySet(55, 65,  75,  85, term="wysoki")
t5 = TrapezoidFuzzySet(75, 85, 100, 100, term="bardzo_wysoki")

FS.add_linguistic_variable("prad_rozladowania",
    LinguisticVariable([t1, t2, t3, t4, t5], universe_of_discourse=[0, 100]))
FS.plot_variable("prad_rozladowania")
```

## 5. napiecie_paneli - Napięcie na panelach solarnych [0–48 V]

```python
t1 = TrapezoidFuzzySet(0,   0,   7,  12, term="bardzo_niskie")
t2 = TrapezoidFuzzySet(7,  12,  19,  24, term="niskie")
t3 = TrapezoidFuzzySet(19, 24,  30,  36, term="srednie")
t4 = TrapezoidFuzzySet(30, 36,  40,  45, term="wysokie")
t5 = TrapezoidFuzzySet(40, 45,  48,  48, term="bardzo_wysokie")

FS.add_linguistic_variable("napiecie_paneli",
    LinguisticVariable([t1, t2, t3, t4, t5], universe_of_discourse=[0, 48]))
FS.plot_variable("napiecie_paneli")
```

## 6. prad_ladowania - Prąd ładowania (wyjście) [0–60 A]

Klasy: bardzo_maly, maly, sredni, duzy, bardzo_duzy

```python
t1 = TrapezoidFuzzySet(0,   0,   9,  15, term="bardzo_maly")
t2 = TrapezoidFuzzySet(9,  15,  21,  27, term="maly")
t3 = TrapezoidFuzzySet(21, 27,  33,  39, term="sredni")
t4 = TrapezoidFuzzySet(33, 39,  45,  51, term="duzy")
t5 = TrapezoidFuzzySet(45, 51,  60,  60, term="bardzo_duzy")

FS.add_linguistic_variable("prad_ladowania",
    LinguisticVariable([t1, t2, t3, t4, t5], universe_of_discourse=[0, 60]))
FS.plot_variable("prad_ladowania")
```

## Reguły systemu Mamdani — 17 reguł

| # | Warunek | Wyjście |
|---|---------|--------|
| 1 | temp_baterii **wysoka** LUB temp_sterownika **wysoka** | prad_ladowania **maly** |
| 2 | temp_baterii **bardzo_wysoka** | prad_ladowania **bardzo_maly** |
| 3 | temp_sterownika **bardzo_wysoka** | prad_ladowania **bardzo_maly** |
| 4 | napiecie_paneli **bardzo_wysokie** | prad_ladowania **bardzo_duzy** |
| 5 | napiecie_paneli **wysokie** | prad_ladowania **duzy** |
| 6 | prad_rozladowania **wysoki** | prad_ladowania **duzy** |
| 7 | prad_rozladowania **bardzo_wysoki** | prad_ladowania **bardzo_duzy** |
| 8 | SOC **bardzo_wysoki** | prad_ladowania **bardzo_maly** |
| 9 | SOC **wysoki** | prad_ladowania **maly** |
| 10 | SOC **sredni** I napiecie_paneli **srednie** | prad_ladowania **sredni** |
| 11 | SOC **niski** I napiecie_paneli **srednie** | prad_ladowania **duzy** |
| 12 | SOC **bardzo_niski** I napiecie_paneli **niskie** | prad_ladowania **duzy** |
| 13 | SOC **bardzo_niski** I napiecie_paneli **bardzo_niskie** | prad_ladowania **sredni** |
| 14 | prad_rozladowania **sredni** I SOC **sredni** | prad_ladowania **sredni** |
| 15 | temp_baterii **niska** I SOC **niski** | prad_ladowania **duzy** |
| 16 | napiecie_paneli **niskie** I SOC **niski** | prad_ladowania **maly** |
| 17 | temp_baterii **srednia** I SOC **sredni** | prad_ladowania **sredni** |

```python
FS.add_rules([
    # R1  — wysoka temperatura baterii LUB sterownika => mały prąd ładowania
    "IF (temp_baterii IS wysoka) OR (temp_sterownika IS wysoka) THEN (prad_ladowania IS maly)",
    # R2  — bardzo wysoka temperatura baterii => bardzo mały prąd (ochrona ogniw)
    "IF (temp_baterii IS bardzo_wysoka) THEN (prad_ladowania IS bardzo_maly)",
    # R3  — bardzo wysoka temperatura sterownika => bardzo mały prąd (ochrona elektroniki)
    "IF (temp_sterownika IS bardzo_wysoka) THEN (prad_ladowania IS bardzo_maly)",
    # R4  — bardzo wysokie napięcie paneli => ładuj maksymalnie
    "IF (napiecie_paneli IS bardzo_wysokie) THEN (prad_ladowania IS bardzo_duzy)",
    # R5  — wysokie napięcie paneli => duży prąd ładowania
    "IF (napiecie_paneli IS wysokie) THEN (prad_ladowania IS duzy)",
    # R6  — wysoki prąd rozładowania => kompensuj dużym ładowaniem
    "IF (prad_rozladowania IS wysoki) THEN (prad_ladowania IS duzy)",
    # R7  — bardzo wysoki prąd rozładowania => kompensuj maksymalnym ładowaniem
    "IF (prad_rozladowania IS bardzo_wysoki) THEN (prad_ladowania IS bardzo_duzy)",
    # R8  — bateria prawie pełna => bardzo mały prąd (ochrona LiFePO4)
    "IF (SOC IS bardzo_wysoki) THEN (prad_ladowania IS bardzo_maly)",
    # R9  — bateria naładowana wysoko => mały prąd
    "IF (SOC IS wysoki) THEN (prad_ladowania IS maly)",
    # R10 — średni SOC, średnie napięcie => środkowy prąd
    "IF (SOC IS sredni) AND (napiecie_paneli IS srednie) THEN (prad_ladowania IS sredni)",
    # R11 — niski SOC, średnie napięcie => duży prąd ładowania
    "IF (SOC IS niski) AND (napiecie_paneli IS srednie) THEN (prad_ladowania IS duzy)",
    # R12 — bardzo niski SOC, niskie napięcie => ładuj ile możesz
    "IF (SOC IS bardzo_niski) AND (napiecie_paneli IS niskie) THEN (prad_ladowania IS duzy)",
    # R13 — bardzo niski SOC, bardzo niskie napięcie => oszczędne ładowanie
    "IF (SOC IS bardzo_niski) AND (napiecie_paneli IS bardzo_niskie) THEN (prad_ladowania IS sredni)",
    # R14 — umiarkowany pobór mocy i SOC => zrównoważony prąd
    "IF (prad_rozladowania IS sredni) AND (SOC IS sredni) THEN (prad_ladowania IS sredni)",
    # R15 — bateria zimna i mało naładowana => duży prąd (niska temp = dobra pojemność)
    "IF (temp_baterii IS niska) AND (SOC IS niski) THEN (prad_ladowania IS duzy)",
    # R16 — niskie napięcie paneli i niski SOC => mały prąd (mało energii)
    "IF (napiecie_paneli IS niskie) AND (SOC IS niski) THEN (prad_ladowania IS maly)",
    # R17 — normalne warunki => standardowy prąd
    "IF (temp_baterii IS srednia) AND (SOC IS sredni) THEN (prad_ladowania IS sredni)",
])
```

## Test systemu

Przykładowe scenariusze:

| Scenariusz | SOC | temp_baterii | temp_sterownika | prad_rozladowania | napiecie_paneli |
|------------|-----|-------------|-----------------|-------------------|----------------|
| Normalny dzień | 50 % | 25 °C | 35 °C | 30 A | 38 V |
| Bateria prawie pełna | 90 % | 28 °C | 40 °C | 10 A | 42 V |
| Upał, wysoki pobór | 40 % | 52 °C | 65 °C | 70 A | 44 V |

```python
scenariusze = [
    {"nazwa": "Normalny dzień",      "SOC": 50, "temp_baterii": 25, "temp_sterownika": 35, "prad_rozladowania": 30, "napiecie_paneli": 38},
    {"nazwa": "Bateria prawie pełna", "SOC": 90, "temp_baterii": 28, "temp_sterownika": 40, "prad_rozladowania": 10, "napiecie_paneli": 42},
    {"nazwa": "Upał, wysoki pobór",   "SOC": 40, "temp_baterii": 52, "temp_sterownika": 65, "prad_rozladowania": 70, "napiecie_paneli": 44},
]

for s in scenariusze:
    FS.set_variable("SOC",               s["SOC"])
    FS.set_variable("temp_baterii",      s["temp_baterii"])
    FS.set_variable("temp_sterownika",   s["temp_sterownika"])
    FS.set_variable("prad_rozladowania", s["prad_rozladowania"])
    FS.set_variable("napiecie_paneli",   s["napiecie_paneli"])

    wynik = FS.inference()
    print(f"[{s['nazwa']}]")
    print(f"  => prad_ladowania = {wynik['prad_ladowania']:.2f} A\n")
```
