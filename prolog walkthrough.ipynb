# Prolog Query Walkthrough

## Query

```prolog
P = para(123, a), (P = para(E, _); P = para(_, E)), \+ number(E).
```

## Step-by-step breakdown

### Step 1 — `P = para(123, a)`

**Unification.** Binds `P` to the compound term `para(123, a)`.

```prolog
P = para(123, a)   % succeeds: P is now bound
```

### Step 2 — `(P = para(E, _) ; P = para(_, E))`

**Disjunction** — Prolog tries both branches left-to-right:

| Branch | Unification | E binds to |
|--------|-------------|------------|
| Left: `P = para(E, _)` | `E = 123`, `_ = a` | **first** argument |
| Right: `P = para(_, E)` | `_ = 123`, `E = a` | **second** argument |

The anonymous variable `_` is a wildcard — it matches anything and is discarded.

### Step 3 — `\+ number(E)`

**Negation-as-failure.** Succeeds only if `E` is **not** a number.

- `\+ number(123)` → **FAILS** (123 is a number)
- `\+ number(a)` → **SUCCEEDS** (`a` is an atom, not a number)

## Execution trace

```
1. P = para(123, a)           → succeeds, P is bound
2. Try left branch: E = 123
   \+ number(123)             → FAILS  ← backtrack!
3. Try right branch: E = a
   \+ number(a)               → SUCCEEDS
```

## Final result

```prolog
E = a.
```

The query finds the **first argument of `para/2` that is not a number** by iterating through both positions via the disjunction and filtering with `\+ number(E)`.

# Example 2 — Recursive `element/2` with Nested `para/2`

## Predicate definitions

```prolog
element(X, para(X, _)).   % X is the first argument of a para/2 term
element(X, para(_, X)).   % X is the second argument of a para/2 term
```

## Query

```prolog
P = para(a, para(b, c)), element(E, P), element(E2, E).
```

## Step 1 — `P = para(a, para(b, c))`

**Unification.** `P` is bound to a nested structure:

```
para
├── a
└── para
    ├── b
    └── c
```

## Step 2 — `element(E, P)` — two branches

Prolog tries both rules against `P = para(a, para(b, c))`:

| Rule | Unification | E |
|------|-------------|---|
| Rule 1: `element(X, para(X, _))` | `X = a`, `_ = para(b,c)` | `E = a` |
| Rule 2: `element(X, para(_, X))` | `_ = a`, `X = para(b,c)` | `E = para(b, c)` |

## Step 3 — `element(E2, E)` — applied to each branch

**Branch `E = a`:**
`element(E2, a)` — `a` is an atom, not a `para(_, _)` term → both rules **fail** → Prolog **backtracks**.

**Branch `E = para(b, c)`:**
`element(E2, para(b, c))` — succeeds twice:

| Rule | Unification | E2 |
|------|-------------|----|
| Rule 1: `element(X, para(X, _))` | `X = b`, `_ = c` | `E2 = b` |
| Rule 2: `element(X, para(_, X))` | `_ = b`, `X = c` | `E2 = c` |

## Execution trace

```
1. P = para(a, para(b, c))          → succeeds

2. element(E, P):
   Try Rule 1 → E = a
     element(E2, a) → FAILS (a is not a para/2 term) ← backtrack
   Try Rule 2 → E = para(b, c)
     element(E2, para(b, c)):
       Try Rule 1 → E2 = b           → SUCCEEDS ✓
       Try Rule 2 → E2 = c           → SUCCEEDS ✓
```

## Final answers

```prolog
E = para(b, c), E2 = b ;
E = para(b, c), E2 = c.
```

`E` always binds to the nested `para(b, c)` — the only element that is itself a `para/2` term and can be passed back into `element/2`. `E2` iterates over its two elements `b` and `c`.
